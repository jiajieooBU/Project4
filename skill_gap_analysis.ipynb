```python
import ast
from collections import Counter

df = pd.read_csv("data/clean_lightcast_job_postings.csv")
df['SKILLS_NAME'] = df['SKILLS_NAME'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith("[") else [])
all_skills = [skill for sublist in df['SKILLS_NAME'] for skill in sublist]
skill_counts = Counter(all_skills)
top_skills = skill_counts.most_common(10)

print(top_skills)
```

```python
top_skills = [skill for skill, count in top_skills]
top_skills
```
